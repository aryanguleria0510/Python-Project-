# Q1: What are the airlines in the dataset, accompanied by their frequencies?

```python
df.head()
```

```python
df['airline'].nunique() # checking how many airlines in the dataset
```

```python
df['airline'].unique()
```

```python
# Showing all the frequenices of the airline

df['airline'].value_counts()
```

```python
# showing all the data in  horizontal bar graph
df['airline'].value_counts(ascending=True).plot.barh(color = ['blue', 'orange'])
plt.title('Airline with Frequencies')
plt.xlabel('Number of flight')
plt.ylabel('Airline')
plt.show()
```

# Q2: Show on the Bar chart representing the departure Time & Arrival Time?

```python
df.head()
```

```python
df['departure_time'].value_counts()  # showing the departure time on flights
```

```python
df['arrival_time'].value_counts()  # showing the arrival time on flights
```

```python
#showing the arrival and departure time in bar graph
plt.figure(figsize = (17,4))
plt.subplot(1,2,1)
plt.bar(df['departure_time'].value_counts().index , df['departure_time'].value_counts().values, color =['r', 'b'])
plt.title('Departure Time')
plt.xlabel('Departure Time')
plt.ylabel('Number of Frequences')
plt.show()

plt.figure(figsize = (8,4))
plt.title('Arrival Time')
plt.xlabel('A Time')
plt.ylabel('Number of Frequencies')
plt.bar(df['arrival_time'].value_counts().index, df['arrival_time'].value_counts().values , color= ['purple' , 'blue'])
plt.show()
```

# Q3: Show Bar Graph representing the the Source City and Destination City?

# Q4: Does the price varies with the Airline?

```python
df.head()
```

```python
# Grouping the airlines and checking their mean price  

df.groupby('airline')['price'].mean()
```

```python
# Draw a categorical Plot showing the mean ticket price for each airline

sns.catplot(x = 'airline', y='price', kind = 'bar', hue='class',
     palette = 'rocket', data = df)
plt.show()
```

# Q5: Does ticket price changed based on the departure time and arrival time?

```python
df.head()
```

```python
# Checking the mean ticket price based on the Departure Times

df.groupby('departure_time')['price'].mean()
```

```python
# Checking the mean ticket price based on the arrival time

df.groupby('arrival_time')['price'].mean()
```

```python

sns.catplot(
    x='departure_time',
    y='price',
    kind='bar',
    hue='departure_time',   # assign colors based on x variable
    legend=False,           # hide duplicate legend
    height=3,
    aspect=2,
    palette='coolwarm',
    data=df
)

plt.show()
```

```python

sns.catplot(
    x='arrival_time',
    y='price',
    kind='bar',
    hue='arrival_time',   # link palette to the x-variable
    legend=False,         # avoid duplicate legend
    height=3,
    aspect=2,
    palette='Set1',
    data=df
)

plt.show()
```

```python
sns.relplot(x = 'arrival_time', y ='price' , data = df, col='departure_time', kind = 'line')
plt.show()
```
